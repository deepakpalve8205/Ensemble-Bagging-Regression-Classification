# Regression

### define x and y for Regression

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

## Preprocessing

```python
x1=Preprocess(x) 
x1
```

## Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.2,random_state=23)
```

## Random Forest Model for Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(random_state=34)
rf.fit(x_train,y_train)
```

## Training Data Evaluation

```python
from evalute import Regression_evaluate
y_pred_train=rf.predict(x_train)

Regression_evaluate(y_pred_train,y_train)
```

## Testing Data Evaluation

```python
y_pred_test=rf.predict(x_test)

Regression_evaluate(y_pred_test,y_test)
```

## *Hyperparameter Tuning*

```python
grid={
    'n_estimators':range(1,250),
    'max_depth':range(1,200),
    'min_samples_split':range(1,20),
    'min_samples_leaf':range(1,20),
    'max_samples':np.arange(0,1,0.1)
}
```

```python
rs=RandomizedSearchCV(rf,param_distributions=grid,cv=3)
rs.fit(x_train,y_train)
```

```python
rs.best_params_
```

```python
rf1=rs.best_estimator_
rf1
```

```python
y_pred_train=rf1.predict(x_train)

Regression_evaluate(y_pred_train,y_train)
```

```python
y_pred_test=rf1.predict(x_test)

Regression_evaluate(y_pred_test,y_test)
```
